```python
# Grouped and aggregated calculation
dt.groupby('Age').PassengerId.count()
# we are counting the passengerid column and asking python to groupall the passenger on the titanic by age groupby('Age')
# Aggregation we need to perform a calculation be it an average or a frequency count then apply it to each subset of a variable another way to think about group calculations is a split apply combine.
```
